Explain why we have to us the Exception class while creating a Custome Exception?

Standardization: By inheriting from the Exception class, your custom exception adheres to a standard interface and behavior that is expected from exceptions in the language. This consistency makes it easier for other developers to understand and handle your custom exception.

Error handling: Exception classes provide a way to handle and propagate errors in a structured manner. They allow you to capture and handle specific types of exceptions separately, enabling you to handle different error scenarios appropriately.

Catching and raising: When you raise an exception, you can catch and handle it using try-except blocks. By inheriting from the Exception class, your custom exception can be caught using a catch block that is designed to handle general exceptions (except for specific cases where more specific handling is required).

Built-in functionality: The Exception class provides useful built-in functionality and attributes that can be utilized in exception handling. For example, exceptions can store error messages, stack traces, and additional information about the error context, which can assist in debugging and error analysis.

Code readability and maintainability: By using the Exception class, you adhere to established conventions and patterns in the programming language. This makes your code more readable and understandable for other developers, enhancing the maintainability and collaboration aspects of your codebase.

What errors are defined in the Airthmetic Error class? Explain any two with an example

The ArithmeticError class in Python is a base class for arithmetic-related exceptions. It serves as a parent class for more specific arithmetic exception classes. Two common exceptions derived from ArithmeticError are ZeroDivisionError and OverflowError

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.")
```

```python
import sys

try:
    big_number = sys.maxsize
    result = big_number * 10
except OverflowError as e:
    print("Error: Arithmetic operation exceeded maximum limit.")
```

```python
import math

try:
    result = math.sqrt(-1)
except ValueError as e:
    print("Error: Invalid operation on a floating-point number.")
```

Why Look up error class is used? Explain with  an example Keyerror and Index error?

The LookupError class in Python is a base class for exceptions that occur when a key or index is not found during a lookup operation. It serves as a parent class for more specific lookup-related exception classes, such as KeyError and IndexError. 

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
except KeyError as e:
    print("Error: Key not found in the dictionary.")
```
